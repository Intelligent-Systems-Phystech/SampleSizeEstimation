```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import lib.utils as ut
import lib.m_models as mm
```

Here we use a predefined function, that generates a sample. In general, any X and y can be used.

```python
name = 'boston'
X, y = ut.DataLoader(name = name)
X, y = ut.preprocess(X, y)
```

Sample size esrimation method returns a dictionary with answer and some intermediate results. For example, in ACC method we can see probability of confidence region contains vector of parameters

```python
res = mm.ACC(X, y, num = 2, progress_bar=lambda x: tqdm(x, leave = False), multiprocess=True)
```

```python
plt.figure(figsize=(12,7))
plt.plot(res['m'], res['E'])
plt.xlabel("Sample size")
plt.title('Sufficient sample size = ' + str(res['m*']))
_ = plt.ylabel("Probaility of confidence region containing model parameters")
```
